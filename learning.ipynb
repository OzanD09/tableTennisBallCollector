```python
x_data = torch.tensor([[1.,2.],[3.,4.]])

w = torch.tensor([[1.0],[2.0]], requires_grad=True)

print(x_data.requires_grad)
print(w.requires_grad)
```

```python
a = torch.tensor(2.0, requires_grad=True)
b = torch.tensor(3.0, requires_grad=True)
x = torch.tensor(4.0, requires_grad=True)

y = a + b
z = x*y

print(z)
print(z.grad_fn)
print(y.grad_fn)
print(a.grad_fn)
```

```python
a = torch.tensor([[1,2],[3,4]])
b = torch.tensor([[20,30],[40,50]])

element_wise_product = a * b
print(element_wise_product)
```

```python
m1 = torch.tensor([[1,2,3],[4,5,6]])
m2 = torch.tensor([[7,8],[9,10],[11,12]])

matrix_product = m1 @ m2
print(matrix_product)
```

```python
score = torch.tensor([[10.,20.,30.],[5.,10.,15.]])

average_scores = score.mean()
average_per_assignment = score.mean(dim=0)
average_per_student = score.mean(dim=1)

print(average_scores)
print(average_per_assignment)
print(average_per_student)
```

```python
scores = torch.tensor([
    [10,0,5,20,1],
    [1,30,2,5,0]
])

best_indices = torch.argmax(scores, dim=1)
print(best_indices)
```
